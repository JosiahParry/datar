```python
df = tibble(
  group=['a'] + ['b']*2 + ['c']*4,
  x=runif(7)
)
df >> group_by(f.group) >> slice_head(n=2) >> ungroup()
```

```python
df >> group_by(f.group) >> slice_head(prop=.5) >> ungroup()
```

```python
# row_number() is 0-based
mtcars >> filter(row_number() == 0) 
```

```python
mtcars >> filter(row_number() == n() - 1)
```

```python
mtcars >> filter(5 <= row_number() < n()) # cannot filter
```

```python
mtcars >> filter(between(row_number(), 5, n()))
```
